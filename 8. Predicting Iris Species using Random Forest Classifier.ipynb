# Random Forest Classifier 

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

### Loading Iris Data  

```python
iris = load_iris()
```

```python
# convert  the dataset into dataframe
iris_df = pd.DataFrame( data = iris.data, columns = iris.feature_names)

iris_df['species'] = iris.target
```

```python
# plotting scatter plot 

sepal_length = iris.data[:, 0]  # First feature: sepal length
sepal_width = iris.data[:, 1]   # Second feature: sepal width

# Create a scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(sepal_length, sepal_width, c=iris.target, cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')

# Add legend
handles, labels = scatter.legend_elements()
plt.legend(handles, iris.target_names, loc = 'lower right', title = 'Classes')
plt.colorbar(label='Species')

plt.show()
```

```python
# splitting the feature and target 

X = iris.data

Y = iris.target
```

### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

### Model Training

```python
model = RandomForestClassifier(n_estimators = 20)

model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
```

```python
confusion_matrix_cm = confusion_matrix(Y_test, Y_pred)

print(confusion_matrix_cm)
```

```python
plt.figure(figsize = (5, 5))
sns.heatmap(confusion_matrix_cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
